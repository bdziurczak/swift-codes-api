```python
import pandas as pd

df = pd.read_csv('./data/Interns_2025_SWIFT_CODES.csv')
df.head(3)
```

## Code indentification

```python
# Create a new DataFrame df1 based on df
df1 = df.copy()

# Add a new column 'ISHQ'
df1['ISHQ'] = df1['SWIFT CODE'].str.endswith('XXX').apply(lambda x: 'H' if x else 'B')

# Create a new DataFrame 'headquarters' with only headquarters records
headquarters = df1[df1['ISHQ'] == 'H']['SWIFT CODE'].str[:8]

#sets used for speed

hq_set = set(headquarters)
df1['HQ SWIFT CODE'] = df1['SWIFT CODE'].str[:8].apply(
    lambda x: x + 'XXX' if x in hq_set else None
)
df1.head(400)
```

## Code Formatting

```python
df1['COUNTRY ISO2 CODE'] = df1['COUNTRY ISO2 CODE'].str.upper()
df1['COUNTRY NAME'] = df1['COUNTRY NAME'].str.upper()
```
